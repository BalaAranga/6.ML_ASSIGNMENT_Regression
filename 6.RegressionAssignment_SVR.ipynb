```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0) 
```

```python
# standardisation
from sklearn.preprocessing import StandardScaler # procedure
sc = StandardScaler() # procedure
x_train =sc.fit_transform(x_train) # fit will do calculation ,transform will tranform data also
x_test =sc.transform(x_test) # here not required fit for x_test , what we used for fit x_train it will take 
```

```python
from sklearn.svm import SVR # procedure1
# kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’
# SVR(kernel="sigmoid",C=0.01) we can use C input to change parameter to get good result
regressor = SVR(kernel="sigmoid",C=500) # procedure2 , procedure1 & procedure2 only require to change for run other algorithm
regressor.fit(x_train,y_train)  
```

```python
weight = regressor.coef_
weight
```

```python
bais = regressor.intercept_
bais
```

```python
y_pred = regressor.predict(x_test) # AI Brain giving output(predicting)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```
